```python
result = []
x = 1
x_1 = 1
n_iterations = 10

def update(x, x_1):
    x = x + x_1
    return x

for i in range(n_iterations):
    x = update(x, x_1)
    x_1 = x
    result.append(x)


plt.plot(result)    
```

## 4.6 Build Model with Multiple Variables

```python

K = 5
r = 1
b = 1
d = 1
c = 1

x = 1
y = 1
n_steps = 100
x_result = []
y_result = []

def update_y():
    y_1 = y - d*y + c*x*y 
    return y_1

def update_x():
    x_1 = x + r*x*(1- (x/K)) - (1-(1/(b*y+1)))*x
    return x_1

for i in range(n_steps):
    next_x = update_x()
    next_y = update_y()
    x_result.append(next_x)
    y_result.append(next_y)
    x = next_x
    y = next_y


plt.plot(x_result)   
plt.plot(y_result)
```
